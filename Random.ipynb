```python
import cv2
import os

IMG_DIR = "//Users/vishrutgupta/Desktop/exCV_COVID/Explainability/Random_100"
img_list = os.listdir(IMG_DIR)
SAVE_DIR = "//Users/vishrutgupta/Desktop/exCV_COVID/Explainability/Random_100_cropped"

for i in img_list:
    img = cv2.imread(os.path.join(IMG_DIR, i))
    
    # Remove 30 pixels from left and top
    cropped = img[30:, 30:]
    
    cv2.imwrite(os.path.join(SAVE_DIR, i), cropped)
print("all images saved.")
```
